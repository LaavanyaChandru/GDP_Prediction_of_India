# Categorise the states into four groups based on the GDP per capita

Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4, where C1 would have the highest per capita GDP and C4, the lowest). 

The quantile values are (0.20,0.5, 0.85, 1), i.e., the states lying between the 85th and the 100th percentile are in C1; those between the 50th and the 85th percentiles are in C2, and so on.

```python
per_capita=gsva["Per Capita GSDP (Rs.)"].sort_values(ascending=False)
per_capita.quantile([0.20,0.5,0.8,1])
```

```python
#converting per capita to data frame

pd.DataFrame(per_capita)
per_cap=per_capita.reset_index()
```

```python
#Categorise the states into four groups based on the GDP per capita , 
#where C1 would have the highest per capita GDP and C4, the lowest).
#The quantile values are (0.20,0.5, 0.85, 1)

C1=[]
C2=[]
C3=[]
C4=[]
j=0
for i in per_cap["Per Capita GSDP (Rs.)"]:
    if per_capita.quantile(1) >= per_cap.iloc[j,1] > per_capita.quantile(0.8):
        st= per_cap.iloc[j,0]
        C1.append(st)
    elif per_capita.quantile(0.8) >= per_cap.iloc[j,1] > per_capita.quantile(0.5):
        st= per_cap.iloc[j,0]
        C2.append(st)
    elif per_capita.quantile(0.5) >= per_cap.iloc[j,1] > per_capita.quantile(0.2):
        st= per_cap.iloc[j,0]
        C3.append(st)
    else:
        st= per_cap.iloc[j,0]
        C4.append(st)
    j=j+1
```
